### Creating a dataframe with the average fitness score through the generations for every combination of mutation, selection and selection

```python
list_name_columns = []
for file in os.listdir(path+'\\attempts'):
    name = file.replace(".csv", "")
    name = name.replace("attempt_", "")
    list_name_columns.append(name)
    
list_name_columns
```

```python
df = pd.DataFrame()

for i in range(len(list_name_columns)):
    file = pd.read_csv(path+'\\attempts\\'+os.listdir(path+'\\attempts')[i], header=None, index_col=0)
    list_aux = file.mean(axis=1) # calculating the average 
    df[list_name_columns[i]]=list_aux
```

```python
import plotly.graph_objects as go
```

```python
# Create traces
fig = go.Figure()

for idx in range(len(list_name_columns)):
    fig.add_trace(go.Scatter(x=df.index, y=df.iloc[:,idx],
                        mode='lines',
                        name=list_name_columns[idx]))


fig.show()
```
