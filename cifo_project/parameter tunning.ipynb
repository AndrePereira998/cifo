```python
import pandas as pd
import os
import shutil
import glob
import csv
import numpy as np
```

```python
#get path
path = os.getcwd()
```

```python
#lista de todos os files dentro do folder attempts
os.listdir(path+'\\PT_')
```

```python
list_name_columns = []
for file in os.listdir(path+'\\PT_'):
    name = file.replace(".csv", "")
    name = name.replace("PT_", "")
    list_name_columns.append(name)
    
list_name_columns
```

```python
df = pd.DataFrame()

for i in range(len(list_name_columns)):
    file = pd.read_csv(path+'\\PT_\\'+os.listdir(path+'\\PT_')[i], header=None, index_col=0)
    list_aux = file.mean(axis=1) # calculating the average 
    df[list_name_columns[i]]=list_aux
```

```python
df
```

```python
import plotly.graph_objects as go
```

```python
# Create traces
fig = go.Figure()

for idx in range(len(list_name_columns)):
    fig.add_trace(go.Scatter(x=df.index, y=df.iloc[:,idx],
                        mode='lines',
                        name=list_name_columns[idx]))


fig.show()
```
